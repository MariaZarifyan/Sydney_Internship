#### Percentage of correct responses for each item in Speeded Naming

```python
data = pd.read_excel("24_06_20_Speeded_naming_results.xlsx")
```

```python
data = data.drop([24354,24355,24356], axis=0)
data["ID"]=data.ID.astype('int64')
len(set(data.ID))
```

```python
data = data[(data.ID != 105) & (data.ID != 126)]
```

```python
len(set(data.ID))
```

```python
df = data[["Stimulus", "Broad_error_1"]]
#df = df.drop([24354,24355,24356], axis=0)
df['Broad_error_1'] = df['Broad_error_1'].str.strip()
len(df)
```

```python
corr = df[df.Broad_error_1 == "correct"]
len(corr)
```

```python
corr.tail()
```

```python
corr = corr.groupby("Stimulus").count().reset_index()
```

```python
corr.columns = ["Stimulus", "Num_of_Corr_resp"]
```

```python
corr
```

```python
corr["PercentageOfCorr"] = round(corr.Num_of_Corr_resp * 100 / 80, 2)
```

```python
corr.to_excel("20_07_13_Percentage_of_corr_SpeededNaming.xlsx")
```
