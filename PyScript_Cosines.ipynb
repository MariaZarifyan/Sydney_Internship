```python
d = {} # key - stimulus, value - list of cos.similarity values for each semantic neighbour
for ind, row in allcos.iterrows():
    d[ind] = row.values.tolist()
```

```python
for key, value in d.items():
    d[key] = [num for num in value if num >= 0.4] #left only values that are no less than 0.4
```

```python
for key, value in d.items():
    d[key] = round(np.mean(value),3)
cosdf = pd.DataFrame(list(d.items()), columns=['Stimulus', 'MeanNearSemNeigh'])
```
